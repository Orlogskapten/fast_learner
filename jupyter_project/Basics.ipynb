# Ask HM / Show HM, most popular ?

We are going to make analysis about HM post.

```python
from csv import reader
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)
print(hn[0:5])
```

```python
headers = hn[0]
hn = hn[1:]
hn[:5]
```

```python
ask_posts = []
show_posts = []
other_posts = []
for row in hn:
    title = row[1]
    title = title.lower()
    if title.startswith("ask hn") == True:
        ask_posts.append(row)
    elif title.startswith("show hn") == True:
        show_posts.append(row)
    else:
        other_posts.append(row)
print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
```

```python
total_ask_comments = 0
for post in ask_posts:
    total_ask_comments += int(post[4])
avg_ask_comments = total_ask_comments / len(ask_posts)
print(avg_ask_comments)

total_show_comments = 0
for row in show_posts:
    total_show_comments += int(row[4])
avg_show_comments = total_show_comments / len(show_posts)
print(avg_show_comments)
```

Ask posts seems to have more comments then Show posts : not anormal because an ask is a question, and s it needs an answer.

**We are going to focus our analysis on Ask posts.**

```python
import datetime as dt

result_list = []
for post in ask_posts:
    date_creation = post[6]
    num_comments = int(post[4])
    result_list.append([date_creation, num_comments])

counts_hours = {}
comments_hours = {}
for row in result_list:
    timer = row[0]
    num = row[1]
    dt_timer = dt.datetime.strptime(timer, "%m/%d/%Y %H:%M")
    dt_hour = dt_timer.hour
    if dt_hour not in counts_hours:
        counts_hours[dt_hour] = 1
        comments_hours[dt_hour] = num
    else:
        counts_hours[dt_hour] += 1
        comments_hours[dt_hour] += num
```

```python
#diviser par count_hours
avg_com_h = []
for hours in counts_hours:
    avg_com_h.append([hours, (comments_hours[hours])/(counts_hours[hours])])
```

```python
avg_com_h
```

```python
swap_avg_hours = []
for hours in avg_com_h:
    swap_avg_hours.append([hours[1],hours[0]])
sorted_swap = sorted(swap_avg_hours, reverse = True)

final_result = []
for value in sorted_swap:
    hours = str(value[1])
    com_avg = float(value[0])
    
    dt_hour = dt.datetime.strptime(hours, "%H")
    st_hour = dt_hour.strftime("%H:%M")
    
    phrase = "{h}: {c:.2f} average comments per post".format(h = st_hour, c = com_avg)
    final_result.append(phrase)
for value in final_result:
    print(value)
```
